## Divide the data into three categories based on fare as below:

fare | category
-|-
0-100  | 'A'
101 - 199 |'B'
200+ |'C'

```python
data.head(2)
```

```python
def test2(x):
    if x <= 100:
        return 'A'
    elif 100 < x < 200:
        return 'B'
    else:
        return 'C'
```

```python
data['Category'] = data['Fare'].apply(test2)
```
